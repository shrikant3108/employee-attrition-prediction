```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score , classification_report
from sklearn.linear_model import LassoCV
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.model_selection import cross_val_score
```

```python
df1 = pd.read_csv("employee_data.csv")
df = df1.copy()

X = df.drop("Employee_Turnover", axis = 1)
Y = df["Employee_Turnover"]

x_train,x_test,y_train,y_test = train_test_split(X,Y, test_size = 0.2 , random_state = 42)

# Losgistic Regration 

model = LogisticRegression (max_iter = 1000)
model.fit(x_train,y_train)


# a = [0.01, 0.1,1,2,3,5,10,20,30,40,50]
    
model1 = LogisticRegression(penalty='l1', solver='liblinear', C=0.5)   # L1 Lasso
model1.fit(x_train,y_train)

model2 = LogisticRegression(penalty='l2', C=1, max_iter=200)   # L2  Ridge
model2.fit(x_train,y_train)

models = {"Losgistic Regration" : model , "L1 Lasso" : model1 , "L2 Ridge" : model2 }

for name , mod in models.items() :
    print(f"Model is : {name}")
    print()
    model_predict_data = mod.predict(x_test)
    print(f"Accurecy of model : {accuracy_score(y_test,model_predict_data)*100} %")
    print()
    cv_accuracy = cross_val_score(mod, X, Y, cv=5, scoring="accuracy")
    cv_f1 = cross_val_score(mod, X, Y, cv=5, scoring="f1")
    
    print("CV Accuracy Scores:", cv_accuracy)
    print()
    print("Mean CV Accuracy:", cv_accuracy.mean()*100 ,"%")
    print()
    print("Mean CV F1 Score:", cv_f1.mean()*100 ,"%")
    print()
    print(f"Classification_Report For Model : {name}")
    print()
    print(classification_report(y_test ,model_predict_data ))
    print("*********************************************************************************************")

    

```
